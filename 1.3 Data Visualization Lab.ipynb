**Hint:**  Import the Spotify data using `df = pd.read_csv('spotify-2023.csv', encoding='ISO-8859-1')`.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('spotify-2023.csv', encoding='ISO-8859-1')
df.head()
```

```python
engineered_songs = df.loc[
    (df['key'] == 'A') &
    (df['bpm'] > 120) &
    (df['bpm'] < 140)
]

engineered_songs
```

```python
vampire = df.loc[df['track_name'] == 'vampire']

vampire
```

```python
df['bpm'].hist(bins=20)
plt.title('Distribution of BPM')
plt.xlabel('BPM')
plt.ylabel('Count')
plt.show()
```

Most songs are clustered in a middle BPM range, so a smash hit would likely have a moderate to fast tempo.


```python
df['key'].value_counts().plot(kind='bar')
plt.title('Distribution of Key')
plt.xlabel('Key')
plt.ylabel('Count')
plt.show()
```

The most common key appears most often in the dataset, so it is a good candidate for a hit song.

```python
df['mode'].value_counts().plot(kind='bar')
plt.title('Distribution of Mode')
plt.xlabel('Mode')
plt.ylabel('Count')
plt.show()
```

The most common mode appears more frequently, so that mode may be more typical of popular songs.

```python
df['danceability_%'].hist(bins=20)
plt.title('Distribution of Danceability')
plt.xlabel('Danceability %')
plt.ylabel('Count')
plt.show()
```

Most songs fall in a middle-to-high danceability range, so a hit song should probably be fairly danceable.

```python
df['energy_%'].hist(bins=20)
plt.title('Distribution of Energy')
plt.xlabel('Energy %')
plt.ylabel('Count')
plt.show()
```

Popular songs seem to have moderate to high energy, so a hit song should likely not be too low-energy.

```python
df['speechiness_%'].hist(bins=20)
plt.title('Distribution of Speechiness')
plt.xlabel('Speechiness %')
plt.ylabel('Count')
plt.show()
```

Most songs appear to have lower speechiness, so a hit song would likely avoid very high speech-like vocals.

```python
df['acousticness_%'].hist(bins=20)
plt.title('Distribution of Acousticness')
plt.xlabel('Acousticness %')
plt.ylabel('Count')
plt.show()
```

Many songs appear to have lower acousticness, so a hit song may lean more produced than acoustic.

```python
print(df['key'].value_counts())
print(df['mode'].value_counts())
print(df[['bpm', 'danceability_%', 'energy_%', 'speechiness_%', 'acousticness_%']].describe())
```

Based on the visualizations and summary statistics, a guaranteed smash hit would most likely be in the key of C# and use Major mode, since those appear most often in the dataset. For the quantitative features, a good range would be BPM between 100 and 140, danceability between 57 and 78, energy between 53 and 77, speechiness between 4 and 11, and acousticness between 6 and 43. These values were chosen because they represent the most typical range for popular songs in the dataset.

```python
engineered_songs = df.loc[
    (df['key'] == 'C#') &
    (df['mode'] == 'Major') &
    (df['bpm'] >= 100) &
    (df['bpm'] <= 140) &
    (df['danceability_%'] >= 57) &
    (df['danceability_%'] <= 78) &
    (df['energy_%'] >= 53) &
    (df['energy_%'] <= 77) &
    (df['speechiness_%'] >= 4) &
    (df['speechiness_%'] <= 11) &
    (df['acousticness_%'] >= 6) &
    (df['acousticness_%'] <= 43)
]

engineered_songs[['track_name', 'artist(s)_name', 'bpm', 'key', 'mode',
                  'danceability_%', 'energy_%', 'speechiness_%', 'acousticness_%',
                  'streams', 'in_spotify_playlists']]
```

The filter returned 2 songs that match the engineered smash hit criteria. These songs appear to be popular because they have high stream counts and are included in many Spotify playlists. This suggests that the selected feature ranges do match characteristics of successful songs in the dataset.

```python
vampire = df.loc[df['track_name'] == 'vampire']
vampire[['track_name', 'artist(s)_name', 'bpm', 'key', 'mode',
         'danceability_%', 'energy_%', 'speechiness_%', 'acousticness_%',
         'streams', 'in_spotify_playlists']]
```
