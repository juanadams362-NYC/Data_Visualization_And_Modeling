```python
import pandas as pd
df = pd.read_csv('actaruslab_f1_telemetry_2026.csv')
df.head()
```

This dataset contains synthetic telemetry data for a Formula 1 car and focuses on aerodynamic stability and porpoising behavior. It was taken from Kaggle and includes 150,000 simulated samples. The data appears to have been generated using a predictive model rather than collected directly from live race telemetry. The main features include speed_kmh, wing_angle_deg, drs_active, downforce_n, drag_n, and stability_index. I chose this dataset because I am interested in cars, racing, and vehicle performance, and this dataset gives a good way to explore how aerodynamic factors affect stability.


1. What is the average speed of the cars in the dataset?


```python
df['speed_kmh'].mean()
```

The average speed in the dataset is shown above in kilometers per hour.

```python
df['stability_index'].mean()
```

The average stability index in the dataset is shown above.

```python
df['drs_active'].value_counts()
```

The output above shows how many samples have DRS inactive (0) and active (1).

```python
df['downforce_n'].max()
```

The maximum downforce recorded in the dataset is shown above in Newtons.

```python
df.groupby('wing_angle_deg')['drag_n'].mean().sort_values(ascending=False).head(10)
```

The output above shows which wing angles are associated with the highest average drag values. This suggests that higher wing angles may increase drag.
